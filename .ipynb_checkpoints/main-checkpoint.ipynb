```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.ndimage import label
from skimage.measure import euler_number
from IPython.display import HTML
from PIL import Image
import librosa

def threshold_image(image, threshold):
    return (image >= threshold).astype(np.uint8)

def animate_filtration(image, num_frames=30, size=None, cmap="gray", interval=600):
    """
    image: 2D NumPy array
    returns: HTML animation for Jupyter/Colab
    """

    image = np.asarray(image)

    # If RGB image, convert to grayscale
    if image.ndim == 3:
        image = image.mean(axis=2)

    # Optional resize
    if size is not None:
        img = Image.fromarray(image.astype(np.uint8))
        img = img.resize(size)
        image = np.array(img)

    thresholds = np.linspace(image.max(), image.min(), num_frames)

    fig, ax = plt.subplots()

    im = ax.imshow(
        np.zeros_like(image),
        cmap=cmap,
        vmin=0,
        vmax=1
    )

    title = ax.set_title("")
    ax.set_xticks([])
    ax.set_yticks([])

    def update(frame):
        t = thresholds[frame]
        binary = threshold_image(image, t)

        _, beta_0 = label(binary)
        chi = euler_number(binary, connectivity=1)
        beta_1 = beta_0 - chi

        im.set_data(binary)
        title.set_text(
            f"Threshold t = {t:.2f}\n"
            f"β₀ components = {beta_0}, β₁ holes = {beta_1}"
        )

        return [im, title]

    ani = FuncAnimation(
        fig,
        update,
        frames=len(thresholds),
        interval=interval,
        blit=False,
        repeat=True
    )

    plt.close(fig)

    return HTML(ani.to_jshtml())
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt



def compute_log_mel_spectrogram(
    audio_path,
    sr=22050,
    n_fft=2048,
    hop_length=512,
    n_mels=128
):
    y, sr = librosa.load(audio_path, sr=sr, mono=True)

    S = librosa.feature.melspectrogram(
        y=y,
        sr=sr,
        n_fft=n_fft,
        hop_length=hop_length,
        n_mels=n_mels,
        power=2.0
    )

    return librosa.power_to_db(S, ref=np.max)

def plot_log_mel_spectrogram(log_S):
    plt.imshow(log_S, origin="lower", aspect="auto", cmap="magma")
    plt.colorbar(label="dB")
    plt.xlabel("Time frame")
    plt.ylabel("Mel frequency bin")
    plt.title("Log-mel spectrogram")
    plt.show()

log_S = compute_log_mel_spectrogram("data/guitar.mp3")
print(log_S.shape)  # (frequency_bins, time_frames)
plot_log_mel_spectrogram(log_S)
```

```python
animate_filtration(log_S, num_frames=30)
```

```python
import gudhi
from gudhi.representations import Landscape
import numpy as np
from skimage.transform import resize

def resize_spectrogram(S, shape=(64, 64)):
    return resize(S, shape, anti_aliasing=True)

def compute_persistence(spectrogram):
    cubical_complex = gudhi.CubicalComplex(top_dimensional_cells=spectrogram)
    cubical_complex.compute_persistence()
    
    h0_diagram = cubical_complex.persistence_intervals_in_dimension(0)
    h1_diagram = cubical_complex.persistence_intervals_in_dimension(1)
    
    
    # Removing infinite deaths, landscapes need finite intervals
    h0_finite = h0_diagram[np.isfinite(h0_diagram[:, 1])]
    h1_finite = h1_diagram[np.isfinite(h1_diagram[:, 1])]
    
    # num_landscapes = number of landscape layers λ1, λ2, ...
    # resolution = number of sample points on the x-axis
    landscape = Landscape(num_landscapes=5, resolution=50)
    
    h0_landscape = landscape.fit_transform([h0_finite])
    h1_landscape = landscape.fit_transform([h1_finite])
    
    return h0_diagram, h1_diagram, h0_landscape, h1_landscape

grid_data = np.random.randint(1, 15, size=(7, 7))
print("Random 7x7 Matrix:")
print(grid_data)

h0_diagram, h1_diagram, h0_landscape, h1_landscape = compute_persistence(grid_data)

print(h0_diagram)
print(h1_diagram)

print("H0 landscape shape:", h0_landscape.shape)
print("H1 landscape shape:", h1_landscape.shape)
#print("H0 landscape vector:\n", h0_landscape)
```

```python
from pathlib import Path
import librosa
import numpy as np
import pandas as pd

def create_music_dataset(genres=("pop", "rock")):
    records = []
    skipped = []

    for label, genre in enumerate(genres):
        genre_path = Path("data/genres") / genre

        for file_path in genre_path.glob("*.wav"):
            try:
                spectrogram = compute_log_mel_spectrogram(file_path)

                records.append({
                    "file": str(file_path),
                    "genre": genre,
                    "label": label,
                    "spectrogram": spectrogram
                })

            except Exception as e:
                print(f"Skipping {file_path}: {e}")
                skipped.append(str(file_path))

    print(f"Loaded {len(records)} files.")
    print(f"Skipped {len(skipped)} files.")

    return pd.DataFrame(records), skipped

genres = ["pop", "metal", "country", "classical"]
df, skipped = create_music_dataset(genres)
df.head()
```

```python
def compute_landscapes(spectrogram):
    _, _, h0_landscape, h1_landscape = compute_persistence(spectrogram)
    return np.concatenate([
        h0_landscape.ravel(),
        h1_landscape.ravel()
    ])

df["landscape"] = df["spectrogram"].apply(compute_landscapes)
print(df.head())
```

```python
def fix_spectrogram_shape(S, target_frames=130):
    # S je oblika (128, broj_vremenskih_frejmova)
    if S.shape[1] > target_frames:
        return S[:, :target_frames]
    elif S.shape[1] < target_frames:
        pad_width = target_frames - S.shape[1]
        return np.pad(S, ((0, 0), (0, pad_width)), mode="constant")
    else:
        return S

df["spectrogram_fixed"] = df["spectrogram"].apply(fix_spectrogram_shape)

print(df["spectrogram_fixed"].iloc[0].shape)
```

```python
import torch
import torch.nn as nn

num_classes = len(genres)

class AudioClassifier(nn.Module):

    def __init__(self, use_topology=False, topo_dim=1000):
        super().__init__()

        self.use_topology = use_topology

        self.cnn = nn.Sequential(
            nn.Conv2d(1, 16, 3),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, 3),
            nn.ReLU(),

            nn.AdaptiveAvgPool2d((4, 4)),
            nn.Flatten()
        )

        self.cnn_summary = nn.Linear(32 * 4 * 4, 128)

        if use_topology:
            self.topo_summary = nn.Sequential(
                nn.Linear(topo_dim, 64),
                nn.ReLU(),
                nn.Dropout(0.5),
            )

            self.classifier = nn.Sequential(
                nn.Linear(128 + 64, 128),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(128, num_classes)
            )

        else:
            self.classifier = nn.Sequential(
                nn.Linear(128, 64),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(64, num_classes)
            )

    def forward(self, spectrogram, landscape=None):
        cnn_features = self.cnn(spectrogram)
        cnn_features = self.cnn_summary(cnn_features)

        if self.use_topology:
            topo_features = self.topo_summary(landscape)
            x = torch.cat([cnn_features, topo_features], dim=1)
        else:
            x = cnn_features

        return self.classifier(x)


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_cnn = AudioClassifier(use_topology=False).to(device)
model_cnn_topo = AudioClassifier(use_topology=True).to(device)

print("Models created.")
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split


def make_loaders(df, use_topology=False, batch_size=16, test_size=0.3, random_state=42):
    train_df, test_df = train_test_split(
        df,
        test_size=test_size,
        random_state=random_state,
        stratify=df["label"]
    )

    X_train_spec = torch.tensor(
        np.stack(train_df["spectrogram_fixed"].values),
        dtype=torch.float32
    ).unsqueeze(1)

    X_test_spec = torch.tensor(
        np.stack(test_df["spectrogram_fixed"].values),
        dtype=torch.float32
    ).unsqueeze(1)

    y_train = torch.tensor(
        train_df["label"].values,
        dtype=torch.long
    )

    y_test = torch.tensor(
        test_df["label"].values,
        dtype=torch.long
    )

    if use_topology:
        landscape_train_np = np.stack(train_df["landscape"].values)
        landscape_test_np = np.stack(test_df["landscape"].values)

        mean = landscape_train_np.mean(axis=0)
        std = landscape_train_np.std(axis=0) + 1e-8

        landscape_train_np = (landscape_train_np - mean) / std
        landscape_test_np = (landscape_test_np - mean) / std

        X_train_topo = torch.tensor(landscape_train_np, dtype=torch.float32)
        X_test_topo = torch.tensor(landscape_test_np, dtype=torch.float32)

        train_dataset = TensorDataset(X_train_spec, X_train_topo, y_train)
        test_dataset = TensorDataset(X_test_spec, X_test_topo, y_test)

        topo_dim = X_train_topo.shape[1]

    else:
        train_dataset = TensorDataset(X_train_spec, y_train)
        test_dataset = TensorDataset(X_test_spec, y_test)

        topo_dim = None

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

    return train_loader, test_loader, topo_dim
```

```python
def evaluate(model, loader, device, use_topology):
    model.eval()

    correct = 0
    total = 0

    with torch.no_grad():

        for batch in loader:

            if use_topology:
                spectrograms, landscapes, labels = batch

                outputs = model(
                    spectrograms.to(device),
                    landscapes.to(device)
                )

                labels = labels.to(device)

            else:
                spectrograms, labels = batch

                outputs = model(
                    spectrograms.to(device)
                )

                labels = labels.to(device)

            predicted = torch.argmax(outputs, dim=1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return correct / total

def train_and_evaluate(
    df,
    use_topology=False,
    epochs=50,
    lr=3e-4,
    device="cuda"
):
    train_loader, test_loader, topo_dim = make_loaders(
        df,
        use_topology=use_topology
    )

    if use_topology:
        model = AudioClassifier(
            use_topology=True,
            topo_dim=topo_dim
        ).to(device)
    else:
        model = AudioClassifier(
            use_topology=False
        ).to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0

        for batch in train_loader:
            optimizer.zero_grad()

            if use_topology:
                spectrograms, landscapes, labels = batch
                spectrograms = spectrograms.to(device)
                landscapes = landscapes.to(device)
                labels = labels.to(device)

                outputs = model(spectrograms, landscapes)

            else:
                spectrograms, labels = batch
                spectrograms = spectrograms.to(device)
                labels = labels.to(device)

                outputs = model(spectrograms)

            loss = criterion(outputs, labels)

            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        if (epoch + 1) % 10 == 0:

            train_acc = evaluate(
                model,
                train_loader,
                device,
                use_topology
            )
        
            test_acc = evaluate(
                model,
                test_loader,
                device,
                use_topology
            )
        
            print(
                f"Epoch {epoch+1}/{epochs} | "
                f"Loss={running_loss/len(train_loader):.4f} | "
                f"Train={train_acc:.4f} | "
                f"Test={test_acc:.4f}"
            )

    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for batch in test_loader:

            if use_topology:
                spectrograms, landscapes, labels = batch
                spectrograms = spectrograms.to(device)
                landscapes = landscapes.to(device)
                labels = labels.to(device)

                outputs = model(spectrograms, landscapes)

            else:
                spectrograms, labels = batch
                spectrograms = spectrograms.to(device)
                labels = labels.to(device)

                outputs = model(spectrograms)

            predicted = torch.argmax(outputs, dim=1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = correct / total

    return model, accuracy
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_cnn, cnn_accuracy = train_and_evaluate(
    df,
    use_topology=False,
    epochs=60,
    lr=3e-4,
    device=device
)

model_pcnn, pcnn_accuracy = train_and_evaluate(
    df,
    use_topology=True,
    epochs=60,
    lr=3e-4,
    device=device
)

print(f"CNN accuracy:  {cnn_accuracy:.4f}")
print(f"PCNN accuracy: {pcnn_accuracy:.4f}")
```
